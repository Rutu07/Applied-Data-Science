```python
## Histogram: Equivalent of numerical variable
```

```python
plt.hist(tips_data['tip'],bins=5)
plt.xlabel('Tips')
```

```python
## Boxplot
```

```python
plt.boxplot(tips_data['total_bill'])
plt.text(x=1.1,y=tips_data['total_bill'].min(),s='min')
plt.text(x=1.1,y=tips_data['total_bill'].quantile(0.25),s='Q1')
plt.text(x=1.1,y=tips_data['total_bill'].median(),s='median(Q2)')
plt.text(x=1.1,y=tips_data['total_bill'].quantile(0.75),s='Q3')
plt.text(x=1.1,y=tips_data['total_bill'].max(),s='max')

plt.title('Box Plot of Total Bill Amount')
plt.ylabel("Total Bill")

plt.show()
```

```python
## Strip Plot: Shows similar to bar plot, shows variation of entire data
```

```python
sns.stripplot(y='tip',x='time',data=tips_data) #Now you know, ore people for dinner, but not necessary high tips
```

```python
sns.stripplot(y='tip',x='time',data=tips_data,jitter=True)
```

```python
## Swarm Plot: Similatr to stripplot, but points are adjused so that they dont overlap
```

```python
sns.swarmplot(y='tip',x='time',data=tips_data,hue='day')
```

```python
## Violin Plot: Similar to swarm plot, except of individual points, it smooths out the distribution. It's symmetric
```

```python
sns.violinplot(y='total_bill',x='day',data=tips_data)
```

```python
sns.violinplot(y='total_bill',x='day',data=tips_data,hue='sex',split=False)
```

```python
sns.violinplot(y='total_bill',x='day',data=tips_data,hue='sex',split=True)
```

```python
## Histogram==> Sample
## Distribution plot==> is an idea of the population
```

```python
sns.distplot(tips_data['total_bill'])
```

```python
time=['Lunch','Dinner']

for i in time:
    subset=tips_data[tips_data['time']==i]
    
    sns.distplot(subset['total_bill'],hist=False,kde=True,
                 kde_kws={'shade':True},
                 label=i)
```
